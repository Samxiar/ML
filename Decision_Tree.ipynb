```python
import pandas as pd
```

```python
df = pd.read_csv("salaries.csv")
df.head()
```

```python
inputs = df.drop("sal_100k", axis="columns")
inputs
```

```python
target=df['sal_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
c_company =LabelEncoder()
j_job =LabelEncoder()
d_degree =LabelEncoder()
```

```python
inputs['company_n'] = c_company.fit_transform(inputs['Company'])
inputs['job_b'] = j_job.fit_transform(inputs['Job'])
inputs['degree_n'] = d_degree.fit_transform(inputs['Degree'])
```

```python
inputs.head()
```

```python
input_n = inputs.drop(['Company','Job','Degree'], axis='columns')
input_n
# abc pharma = 0
# Facebook = 1
# Google = 2
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```

```python
model.score(input_n,target)
```

```python
import warnings
warnings.simplefilter("ignore")
model.predict([['2','2','1']])
```
